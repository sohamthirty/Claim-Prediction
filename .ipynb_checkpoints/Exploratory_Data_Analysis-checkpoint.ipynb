# Exploratory Data Analysis

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

## Import Dataset

```python
df = pd.read_csv("Data/travel_insurance.csv")
```

## EDA

### Numerical Statistics

```python
#Check data shape 
#(num of Rows & Columns)
df.shape
```

```python
#Check each data type of columns and missing values
df.info()
```

```python
#Check the percentages of missing value
df.isnull().sum() / df.shape[0] * 100
```

```python
#Summary Statistics
df.describe()
```

```python
#Check value counts for all columns
Categorical_features = ['Agency', 'Agency Type', 'Distribution Channel', 'Product Name','Claim', 
                        'Destination', 'Gender']

for i in Categorical_features:
    print(df[i].value_counts())
    print("----------------------------------------------------------------------")
```

### Data Visualization

#### Countplots

```python
# Countplot for 'Claim' variable
fig = sns.countplot(x = df['Claim'])

fig.text(x=1,y=6000,s='Claimed % = 1.4',fontsize=16,weight='bold',ha='center',va='bottom',color='navy')
fig.set_title('Claim', fontsize=20)
```

```python
# Countplot for 'Agency Type' variable
fig = sns.countplot(x = df['Agency Type'])
fig.set_title('Agency Type', fontsize=20)
```

```python
# Countplot for 'Agency' variable
from matplotlib.pyplot import figure

plt.figure(figsize=(14, 6))
fig = sns.countplot(x = df['Agency'])
fig.set_title('Agency', fontsize=20)
```

```python
fig = sns.countplot(x = df['Distribution Channel'])
fig.set_title('Distribution Channel', fontsize=20)
```

```python
plt.figure(figsize=(18,8))
fig = sns.countplot(x = df['Destination'])
fig.set_title('Destination', fontsize=20)
h = fig.get_xticklabels()
k = fig.set_xticklabels(h,rotation = 80)
```

```python
plt.figure(figsize=(18,8))
fig = sns.countplot(x = df['Product Name'])
fig.set_title('Product Names', fontsize=20)
h = fig.get_xticklabels()
k = fig.set_xticklabels(h,rotation = 80)
```

#### Correlation Matrix

```python
#Check the correlation between variables in the data
corrMatrix = df.corr()

#Visualise the matrix with Seaborn library
plt.figure(figsize=(6,2))
sns.heatmap(corrMatrix, annot = True, cmap= 'coolwarm')
```
